```python
import database
import pandas as pd

df = database.load_all_leads()
```

```python
print(df['DOR_UC'].value_counts().head(15))
```

```python
print(f"starting records: {len(df):,}")
print(f"columns: {list(df.columns)}")
print(f"\ndata types:")
print(df.dtypes)
```

```python
print("=" * 50)
print("NULL AND ZERO ANALYSIS")
print("=" * 50)

# Check actual nulls
print("\n--- ACTUAL NULL VALUES ---")
nulls = df.isnull().sum()
print(nulls[nulls > 0])

# Check zeros in numeric columns
print("\n--- ZERO VALUES IN KEY COLUMNS ---")
numeric_cols = [
    'JV', 'ACT_YR_BLT', 'EFF_YR_BLT',
    'TOT_LVG_AR', 'SALE_PRC1', 'SALE_PRC2',
    'LND_SQFOOT', 'NO_BULDNG', 'JV_HMSTD'
]
for col in numeric_cols:
    zeros = (df[col] == 0).sum()
    nulls_in_col = df[col].isnull().sum()
    print(f"{col:<15} zeros: {zeros:>5,}  |  nulls: {nulls_in_col:>5,}")

# Check empty strings in text columns
print("\n--- EMPTY STRINGS IN TEXT COLUMNS ---")
text_cols = ['PHY_CITY', 'OWN_NAME', 'OWN_STATE_', 'QUAL_CD1', 'NBRHD_CD']
for col in text_cols:
    empty = (df[col].astype(str).str.strip() == '').sum()
    print(f"{col:<15} empty strings: {empty:>5,}")
```

```python
# Step 1 — Residential Filter
# Normalize DOR_UC first — strip whitespace, zero-pad to 3 digits
df['DOR_UC'] = (
    df['DOR_UC']
    .astype(str)    # convert to string in case any stored as number
    .str.strip()    # remove any leading/trailing spaces
    .str.zfill(3)   # pad to 3 digits — turns '1' into '001'
)

RESIDENTIAL_CODES = ['001', '002', '007', '010']

before = len(df)
df_clean = df[df['DOR_UC'].isin(RESIDENTIAL_CODES)].copy()
after = len(df_clean)

print(f"Before filter : {before:,}")
print(f"After filter  : {after:,}")
print(f"Removed       : {before - after:,} non-residential records")
print(f"\nRemaining DOR codes:")
print(df_clean['DOR_UC'].value_counts())
```

```python
# Step 2 — Remove invalid JV values
before = len(df_clean)

df_clean = df_clean[df_clean['JV'] >= 1000].copy()

after = len(df_clean)

print(f"Before JV filter : {before:,}")
print(f"After JV filter  : {after:,}")
print(f"Removed          : {before - after:,} records with JV below $1,000")
print(f"\nJV statistics after removal:")
print(f"Min JV  : ${df_clean['JV'].min():>12,.0f}")
print(f"Max JV  : ${df_clean['JV'].max():>12,.0f}")
print(f"Mean JV : ${df_clean['JV'].mean():>12,.0f}")
print(f"Median JV:${df_clean['JV'].median():>12,.0f}")
```

```python
# Step 3 — Remove JV outliers using IQR method
import numpy as np

before = len(df_clean)

# Calculate the quartiles
Q1  = df_clean['JV'].quantile(0.25)
Q3  = df_clean['JV'].quantile(0.75)
IQR = Q3 - Q1

# Calculate the upper boundary
upper_bound = Q3 + (1.5 * IQR)

print(f"Q1  (25th percentile) : ${Q1:>12,.0f}")
print(f"Q3  (75th percentile) : ${Q3:>12,.0f}")
print(f"IQR (Q3 minus Q1)     : ${IQR:>12,.0f}")
print(f"Upper bound           : ${upper_bound:>12,.0f}")
print(f"")

# Apply the filter — keep only records below upper bound
df_clean = df_clean[df_clean['JV'] <= upper_bound].copy()

after = len(df_clean)

print(f"Before outlier removal : {before:,}")
print(f"After outlier removal  : {after:,}")
print(f"Removed                : {before - after:,} extreme outliers")
print(f"")
print(f"JV statistics after outlier removal:")
print(f"Min JV   : ${df_clean['JV'].min():>10,.0f}")
print(f"Max JV   : ${df_clean['JV'].max():>10,.0f}")
print(f"Mean JV  : ${df_clean['JV'].mean():>10,.0f}")
print(f"Median JV: ${df_clean['JV'].median():>10,.0f}")
```

```python
# Step 4 — Standardize and fill PHY_CITY

# ZIP to city mapping for Alachua County
ZIP_TO_CITY = {
    32601: 'GAINESVILLE', 32602: 'GAINESVILLE',
    32603: 'GAINESVILLE', 32604: 'GAINESVILLE',
    32605: 'GAINESVILLE', 32606: 'GAINESVILLE',
    32607: 'GAINESVILLE', 32608: 'GAINESVILLE',
    32609: 'GAINESVILLE', 32610: 'GAINESVILLE',
    32611: 'GAINESVILLE', 32612: 'GAINESVILLE',
    32641: 'GAINESVILLE', 32653: 'GAINESVILLE',
    32615: 'ALACHUA',     32616: 'ALACHUA',
    32694: 'WALDO',       32643: 'HIGH SPRINGS',
    32640: 'HAWTHORNE',   32666: 'KEYSTONE HEIGHTS',
    32668: 'MICANOPY',
}

CITY_CORRECTIONS = {
    'LACROSSE' : 'LA CROSSE',
    'LA CROSSE': 'LA CROSSE',
}

# Step 4a — Strip whitespace and uppercase everything
df_clean['PHY_CITY'] = (
    df_clean['PHY_CITY']
    .astype(str)
    .str.strip()
    .str.upper()
)

# Step 4b — Replace 'NAN' string that appears after astype(str)
df_clean['PHY_CITY'] = df_clean['PHY_CITY'].replace('NAN', '')

# Step 4c — Apply known spelling corrections
df_clean['PHY_CITY'] = df_clean['PHY_CITY'].replace(CITY_CORRECTIONS)

# Step 4d — Count empty before filling
empty_before = (df_clean['PHY_CITY'] == '').sum()

# Step 4e — Fill empty cities from ZIP lookup
# First convert PHY_ZIPCD to integer for dictionary lookup
df_clean['PHY_ZIPCD'] = df_clean['PHY_ZIPCD'].fillna(0).astype(int)

empty_mask = df_clean['PHY_CITY'] == ''
df_clean.loc[empty_mask, 'PHY_CITY'] = (
    df_clean.loc[empty_mask, 'PHY_ZIPCD']
    .map(ZIP_TO_CITY)
    .fillna('UNKNOWN')
)

empty_after = (df_clean['PHY_CITY'] == '').sum()
unknown_count = (df_clean['PHY_CITY'] == 'UNKNOWN').sum()

print(f"Empty cities before   : {empty_before:,}")
print(f"Filled from ZIP       : {empty_before - empty_after - unknown_count:,}")
print(f"Could not map (UNKNOWN): {unknown_count:,}")
print(f"Empty cities after    : {empty_after:,}")
print(f"")
print(f"City distribution after cleaning:")
print(df_clean['PHY_CITY'].value_counts())
```

```python
# Step 5 — Impute zero ACT_YR_BLT

print("Before imputation:")
print(f"Zero ACT_YR_BLT : {(df_clean['ACT_YR_BLT'] == 0).sum():,}")
print(f"Valid ACT_YR_BLT: {(df_clean['ACT_YR_BLT'] > 0).sum():,}")
print()

# Median year per DOR code — only from valid records
group_medians = (
    df_clean[df_clean['ACT_YR_BLT'] > 0]
    .groupby('DOR_UC')['ACT_YR_BLT']
    .median()
)
print("Median year by DOR group:")
print(group_medians)
print()

# Overall median as fallback
overall_median = int(
    df_clean[df_clean['ACT_YR_BLT'] > 0]['ACT_YR_BLT'].median()
)
print(f"Overall median year (fallback): {overall_median}")
print()

# Apply imputation
zero_mask = df_clean['ACT_YR_BLT'] == 0
df_clean.loc[zero_mask, 'ACT_YR_BLT'] = (
    df_clean.loc[zero_mask, 'DOR_UC']
    .map(group_medians)
    .fillna(overall_median)
)
df_clean['ACT_YR_BLT'] = df_clean['ACT_YR_BLT'].astype(int)

# Fix EFF_YR_BLT zeros — set equal to ACT_YR_BLT where missing
eff_zero_mask = df_clean['EFF_YR_BLT'] == 0
df_clean.loc[eff_zero_mask, 'EFF_YR_BLT'] = (
    df_clean.loc[eff_zero_mask, 'ACT_YR_BLT']
)
df_clean['EFF_YR_BLT'] = df_clean['EFF_YR_BLT'].astype(int)

print("After imputation:")
print(f"Zero ACT_YR_BLT remaining : {(df_clean['ACT_YR_BLT'] == 0).sum():,}")
print(f"Zero EFF_YR_BLT remaining : {(df_clean['EFF_YR_BLT'] == 0).sum():,}")
print(f"Year range : {df_clean['ACT_YR_BLT'].min()} to {df_clean['ACT_YR_BLT'].max()}")
```

```python
# Step 6 — Flag never sold properties

print("Before flagging:")
print(f"SALE_PRC1 = 0 : {(df_clean['SALE_PRC1'] == 0).sum():,}")
print(f"SALE_PRC1 > 0 : {(df_clean['SALE_PRC1'] > 0).sum():,}")
print()

# Create never_sold flag
# Zero sale price means never sold or transferred for $0
# This is INFORMATION not missing data — do not remove
df_clean['never_sold'] = (df_clean['SALE_PRC1'] == 0).astype(int)

print("After flagging:")
print(f"never_sold = 1 : {df_clean['never_sold'].sum():,}  (never transacted)")
print(f"never_sold = 0 : {(df_clean['never_sold'] == 0).sum():,}  (has sale record)")
print()
print(f"Percentage never sold: {df_clean['never_sold'].mean() * 100:.1f}%")
```

```python
# Step 7 — Handle TOT_LVG_AR zeros
import numpy as np

print("Before fixing:")
print(f"TOT_LVG_AR = 0   : {(df_clean['TOT_LVG_AR'] == 0).sum():,}")
print(f"TOT_LVG_AR > 0   : {(df_clean['TOT_LVG_AR'] > 0).sum():,}")
print(f"TOT_LVG_AR null  : {df_clean['TOT_LVG_AR'].isnull().sum():,}")
print()

# Zero living area means vacant land or data not recorded
# Cannot calculate price per sqft on zero sqft
# Set to NaN — feature engineering will handle it
zero_mask = df_clean['TOT_LVG_AR'] == 0
df_clean.loc[zero_mask, 'TOT_LVG_AR'] = np.nan

print("After fixing:")
print(f"TOT_LVG_AR = 0   : {(df_clean['TOT_LVG_AR'] == 0).sum():,}")
print(f"TOT_LVG_AR > 0   : {(df_clean['TOT_LVG_AR'] > 0).sum():,}")
print(f"TOT_LVG_AR null  : {df_clean['TOT_LVG_AR'].isnull().sum():,}")
print()
print(f"Valid living area stats:")
print(f"Min  : {df_clean['TOT_LVG_AR'].min():>8,.0f} sqft")
print(f"Max  : {df_clean['TOT_LVG_AR'].max():>8,.0f} sqft")
print(f"Mean : {df_clean['TOT_LVG_AR'].mean():>8,.0f} sqft")
```

```python
# Step 8 — Save clean dataset to database as new table

import sqlite3
import project_config as config

print(f"Records ready to save : {len(df_clean):,}")
print(f"Columns in clean data : {len(df_clean.columns)}")
print(f"Columns: {list(df_clean.columns)}")
print()

# Save to SQLite as a new table called leads_clean
# if_exists='replace' means if the table already exists
# it gets dropped and recreated with fresh data
conn = sqlite3.connect(config.DB_NAME)
df_clean.to_sql('leads_clean', conn, if_exists='replace', index=False)
conn.close()

print("Saved successfully.")
print()

# Verify by reloading from database
conn = sqlite3.connect(config.DB_NAME)
df_verify = pd.read_sql_query("SELECT COUNT(*) as total FROM leads_clean", conn)
conn.close()

print(f"Verification — records in leads_clean table: {df_verify['total'][0]:,}")
print()
print("Cleaning pipeline complete.")
print(f"Original raw table  : leads       → 8,000 records  (untouched)")
print(f"Clean output table  : leads_clean → {len(df_clean):,} records")
```
