```python
import database
import pandas as pd
import matplotlib.pyplot as plt

df = database.load_all_leads()

# ─── CHECK 1: How many records have zero sale price ───
zero_sales = (df['SALE_PRC1'] == 0).sum()
valid_sales = (df['SALE_PRC1'] > 0).sum()
print(f"Records with ZERO sale price: {zero_sales:,} ({zero_sales/len(df)*100:.1f}%)")
print(f"Records with VALID sale price: {valid_sales:,} ({valid_sales/len(df)*100:.1f}%)")
print()
```

```python
# ─── CHECK 2: Year built distribution ───
print("YEAR BUILT DISTRIBUTION:")
print(f"  Zero/Unknown: {(df['ACT_YR_BLT'] == 0).sum():,}")
print(f"  Valid years:  {(df['ACT_YR_BLT'] > 0).sum():,}")
print(f"  Year range:   {df[df['ACT_YR_BLT'] > 0]['ACT_YR_BLT'].min():.0f} to {df[df['ACT_YR_BLT'] > 0]['ACT_YR_BLT'].max():.0f}")
print()
```

```python
# ─── CHECK 3: Property type distribution ───
print("PROPERTY TYPE (DOR_UC) DISTRIBUTION:")
print(df['DOR_UC'].value_counts().head(15))
print()
```

```python
# ─── CHECK 4: Homestead distribution ───
homestead = (df['JV_HMSTD'] > 0).sum()
no_homestead = (df['JV_HMSTD'] == 0).sum()
print(f"Owner occupied (homestead): {homestead:,} ({homestead/len(df)*100:.1f}%)")
print(f"Investment/vacant:          {no_homestead:,} ({no_homestead/len(df)*100:.1f}%)")
print()
```

```python
# ─── CHECK 5: Value distribution ───
print("JUST VALUE (JV) DISTRIBUTION:")
print(f"  Zero values:  {(df['JV'] == 0).sum():,}")
print(f"  Under $50k:   {((df['JV'] > 0) & (df['JV'] < 50000)).sum():,}")
print(f"  $50k-$200k:   {((df['JV'] >= 50000) & (df['JV'] < 200000)).sum():,}")
print(f"  $200k-$500k:  {((df['JV'] >= 200000) & (df['JV'] < 500000)).sum():,}")
print(f"  Over $500k:   {(df['JV'] >= 500000).sum():,}")
print()
```

```python
# ─── CHECK 6: Sale year distribution ───
print("SALE YEAR DISTRIBUTION:")
valid_years = df[df['SALE_YR1'] > 0]['SALE_YR1']
print(f"  Never sold (year=0): {(df['SALE_YR1'] == 0).sum():,}")
print(f"  Sold before 2000:    {((df['SALE_YR1'] > 0) & (df['SALE_YR1'] < 2000)).sum():,}")
print(f"  Sold 2000-2015:      {((df['SALE_YR1'] >= 2000) & (df['SALE_YR1'] < 2015)).sum():,}")
print(f"  Sold 2015-2020:      {((df['SALE_YR1'] >= 2015) & (df['SALE_YR1'] < 2020)).sum():,}")
print(f"  Sold 2020-present:   {(df['SALE_YR1'] >= 2020).sum():,}")
```
